```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/Users/archishman/dissertation_project/customer_reviews_qna_agent/data/train.csv')
```

```python
df.columns = ['sentiment_label','review_title','review_text']
```

```python
df['sentiment'] = pd.Series(np.where(df['sentiment_label']==1,'negative','positive'))
```

```python
new_df = df.loc[:49999,:]
```

```python
region_list = np.array(['region1','region2','region3','region4','region5'])
new_df['region'] = pd.Series(np.random.choice(a = region_list,size=len(new_df)))
```

```python
store_list = np.array([f'store{i}' for i in range(1,251)])
new_df['store'] = pd.Series(np.random.choice(a = store_list,size=len(new_df)))
```

```python
final_df = new_df[['review_text','sentiment','region','store']]
```

```python
final_df.to_csv('/Users/archishman/dissertation_project/customer_reviews_qna_agent/data/processed_reviews.csv',index=False)
```
